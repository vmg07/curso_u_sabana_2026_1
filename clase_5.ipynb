```python
# Bloque 1: Creando nuestra primera clase (Diapositiva 5)
# La anatomía de una clase y el método __init__.

# 1. Definimos el "plano" para cualquier producto
class Producto:
    
    # 2. El CONSTRUCTOR: esta función se ejecuta automáticamente al crear un nuevo objeto.
    #    'self' se refiere al objeto específico que se está creando.
    def __init__(self, nombre_producto, precio_producto, stock_inicial):
        
        # 3. ATRIBUTOS: Guardamos los datos que nos pasaron en el objeto 'self'.
        #    Estas son las características del producto.
        self.nombre = nombre_producto
        self.precio = precio_producto
        self.stock = stock_inicial
        
        print(f"¡Se ha creado el producto '{self.nombre}'!")
```

```python
# Bloque 2: Instanciación de Objetos (Diapositiva 6)
# Demostrar cómo dar vida a los objetos.

# Creamos un objeto (instancia) de la clase Producto
cafe_tostao = Producto(nombre_producto="Café de Origen Tostao", precio_producto=15000, stock_inicial=100)

# Creamos OTRO objeto de la misma clase, con datos diferentes
chocolate_jet = Producto(nombre_producto="Chocolatina Jet", precio_producto=500, stock_inicial=500)

# ¿Cómo accedemos a sus atributos? Usando la sintaxis del punto.
print(f"El precio del café es: ${cafe_tostao.precio}") # Salida: 15000
print(f"El stock de Jet es: {chocolate_jet.stock} unidades") # Salida: 500
```

```python
# Bloque 3: Agregando Métodos a la Clase (Diapositiva 8)
# Aquí se redefine la clase Producto agregando los métodos vender y mostrar_info.

class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    # MÉTODO 1: Vender unidades de este producto
    def vender(self, cantidad):
        if self.stock >= cantidad:
            self.stock -= cantidad # Actualiza su propio stock
            print(f"¡Venta exitosa! Se vendieron {cantidad} de {self.nombre}.")
        else:
            print(f"¡Venta fallida! No hay suficiente stock de {self.nombre}.")

    # MÉTODO 2: Mostrar el estado actual del producto
    def mostrar_info(self):
        print(f"--- Info de {self.nombre} ---")
        print(f"Precio: ${self.precio:,.0f}")
        print(f"Stock disponible: {self.stock} unidades")
```

```python
# Bloque 4: Usando los Métodos (Diapositiva 9)
# Simular la interacción con el objeto.

# Creamos nuestro objeto café
cafe_tostao = Producto("Café de Origen Tostao", 15000, 100)

# Usamos sus métodos (acciones)
cafe_tostao.mostrar_info() # Muestra el estado inicial
cafe_tostao.vender(10)     # Realizamos una venta
cafe_tostao.vender(95)     # Intentamos una venta que fallará
cafe_tostao.mostrar_info() # Mostramos el estado final
```

```python
# Bloque 5: Solución al Taller Práctico - Cliente Bancario (Diapositivas 11 a 14)

class ClienteBancario:
    """
    Esta clase es el "plano" o "molde" para crear objetos de tipo cliente bancario.
    Define las características (atributos) y las acciones (métodos) que
    todo cliente tendrá.
    """

    # --- El Constructor: La función de "nacimiento" de cada objeto ---
    def __init__(self, nombre_cliente, cedula_cliente):
        """
        Este método especial se ejecuta automáticamente cada vez que creamos un
        nuevo cliente. Su trabajo es inicializar los atributos del objeto.
        'self' se refiere al objeto específico que se está creando en ese momento.
        """
        # --- Atributos ---
        # Guardamos los datos que nos pasaron como las características del objeto.
        self.nombre = nombre_cliente
        self.cedula = cedula_cliente
        # Todo cliente nuevo empieza con un saldo de 0. Usamos float por ser dinero.
        self.saldo_cuenta = 0.0

        print(f"✅ ¡Bienvenido, {self.nombre}! Su cuenta ha sido creada con el saldo inicial de ${self.saldo_cuenta:,.0f}.")

    # --- Métodos: Las "acciones" o "poderes" que tiene el objeto ---
    def consignar(self, monto):
        """
        Aumenta el saldo de la cuenta. Modifica el estado interno del objeto.
        """
        # Es una buena práctica validar que el monto sea positivo.
        if monto > 0:
            self.saldo_cuenta += monto  # Abreviatura de: self.saldo_cuenta = self.saldo_cuenta + monto
            print(f"💰 Consignación exitosa. Se han añadido ${monto:,.0f} a su cuenta.")
        else:
            print("❌ Error: El monto a consignar debe ser un valor positivo.")

    def retirar(self, monto):
        """
        Disminuye el saldo de la cuenta, pero solo si hay fondos suficientes.
        Aquí es donde el objeto toma una "decisión" basada en su propio estado.
        """
        # Validamos que el monto a retirar sea positivo.
        if monto > 0:
            # La condición clave: el objeto revisa su propio saldo.
            if monto <= self.saldo_cuenta:
                self.saldo_cuenta -= monto
                print(f"💸 Retiro exitoso. Se han debitado ${monto:,.0f} de su cuenta.")
            else:
                # El bloque 'else' se activa si no hay fondos suficientes.
                print(f"🚫 Fondos insuficientes. Intenta retirar ${monto:,.0f}, pero su saldo es de solo ${self.saldo_cuenta:,.0f}.")
        else:
            print("❌ Error: El monto a retirar debe ser un valor positivo.")

    def consultar_saldo(self):
        """
        Simplemente muestra el estado actual del atributo 'saldo_cuenta'.
        """
        print(f"📄 El saldo actual de la cuenta de {self.nombre} es: ${self.saldo_cuenta:,.0f}")


# --- INICIANDO SIMULACIÓN BANCARIA REQUERIDA EN EL TALLER ---

if __name__ == "__main__":
    # 1. Creamos un objeto (una instancia) de la clase ClienteBancario.
    #    Al hacer esto, se llama automáticamente al método __init__.
    print("\nPASO 1: Creando el objeto cliente...")
    mi_cliente = ClienteBancario(nombre_cliente="Diego Zuluaga", cedula_cliente="123456789")

    # 2. Le consignamos $500,000.
    print("\nPASO 2: Consignando $500,000...")
    mi_cliente.consignar(500000)
    mi_cliente.consultar_saldo()

    # 3. Intentamos retirar $600,000 (esta operación debe fallar).
    print("\nPASO 3: Intentando retirar $600,000 (debe fallar)...")
    mi_cliente.retirar(600000)

    # 4. Retiramos $100,000 (esta operación debe ser exitosa).
    print("\nPASO 4: Retirando $100,000 (debe ser exitoso)...")
    mi_cliente.retirar(100000)

    # 5. Consultamos el saldo final para verificar el resultado.
    print("\n--- CONSULTA DE SALDO FINAL ---")
    mi_cliente.consultar_saldo()
```

```python
# Definición de la clase ClienteBancario
class ClienteBancario:
    """
    Clase que representa a un cliente bancario con funcionalidades básicas de cuenta.
    """
    def __init__(self, nombre, cedula):
        """
        Constructor de la clase. Inicializa los atributos del cliente.
        Args:
            nombre (str): Nombre del cliente.
            cedula (str): Cédula del cliente.
        """
        self.nombre = nombre
        self.cedula = cedula
        self.saldo_cuenta = 0.0  # Saldo inicializado en 0
    def consignar(self, monto):
        """
        Método para consignar dinero a la cuenta.
        Args:
            monto (float): Monto a consignar.
        """
        if monto > 0:
            self.saldo_cuenta += monto
            print(f"Consignación exitosa. Nuevo saldo: ${self.saldo_cuenta:,.0f}")
        else:
            print("Error: El monto a consignar debe ser positivo.")
    def retirar(self, monto):
        """
        Método para retirar dinero de la cuenta, solo si hay fondos suficientes.
        Args:
            monto (float): Monto a retirar.
        """
        if monto > 0:
            if monto <= self.saldo_cuenta:
                self.saldo_cuenta -= monto
                print(f"Retiro exitoso. Nuevo saldo: ${self.saldo_cuenta:,.0f}")
            else:
                print("Fondos insuficientes")
        else:
            print("Error: El monto a retirar debe ser positivo.")
    def consultar_saldo(self):
        """
        Método para consultar e imprimir el saldo actual de la cuenta.
        """
        print(f"Saldo actual de {self.nombre}: ${self.saldo_cuenta:,.0f}")
 
# Creación de un objeto de prueba
cliente_prueba = ClienteBancario("Juan Pérez", "123456789")
 
# Operaciones de prueba
print("Consignando $500000...")
cliente_prueba.consignar(500000)
 
print("\nIntentando retirar $600000...")
cliente_prueba.retirar(600000)
 
print("\nRetirando $100000...")
cliente_prueba.retirar(100000)
 
print("\nConsultando saldo final...")
cliente_prueba.consultar_saldo()
```
